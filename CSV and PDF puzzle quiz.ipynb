```python
import PyPDF2

f = open('Find_the_Phone_Number.pdf','rb')
pdf_reader = PyPDF2.PdfFileReader(f)
pdf_reader.numPages

```

```python
import re
```

# this is the critical bit of code that changes the text from binary to readable text.

```python
all_text = ''

for n in range (pdf_reader.numPages):
    page = pdf_reader.getPage(n)
    page_text = page.extractText()
    
    all_text = all_text+' '+page_text
```

```python
all_text
```

```python
pattern = r'\d{3}.\d{3}.\d{4}'
```

```python
# for match in re.finditer(pattern,pdf_reader):
#     print(match)


results = re.search(pattern,all_text)
```
